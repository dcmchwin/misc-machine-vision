# Import data

```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
```

```python
test_img_path = 'shapes_and_colors.jpg'
im = cv2.imread(test_img_path)
gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

_, ax = plt.subplots(1, 2)
ax[0].imshow(im)
ax[1].imshow(gray, cmap='Greys')
plt.show()
```

## Show opencv out-of-the-box method

Important parameters appear to be:

block_size - size of neighbourhood considered for corner detection

ksize - Aperture parameter of Sobel derivative used

k - Harris detector free parameter in the equation

```python
ksize = 3
block_size = 2
k = 0.04
dst = cv2.cornerHarris(gray, block_size, ksize, k)
dst = cv2.dilate(dst, None, iterations=3)

corners = np.zeros(gray.shape)
corners[dst > 0.01 * dst.max()] = 1

masked_data = np.ma.masked_where(corners==0, corners)

plt.figure(figsize=(10,8))
plt.imshow(gray, cmap='Greys')
plt.imshow(masked_data, interpolation='none')
plt.show()
```
